#### B. Creates a new *column* called 'missing_values' in the input DF that sums the missing values in each row.

### Notes: 
***Function takes one argument: An input DataFrame***
<br> </br>
<br> </br>
Review the Column creation method in lecture and the axis arguments needed

### Expected Output:
<img src= 'missingrows_expectedoutput.png'>

```python
#Answer 3B here; Please comment each part



def Missing_values(input_df):
    # Created a dataframe called missing_values
    ## Finding rows of input_df that is equal to null with isnull() function
    ## Then using sum() funciton by rows (using argument axis = 1) to find number of nulls
    ## Then naming column as 'Missing' with columns argument
    missing_values = pd.DataFrame(input_df.isnull().sum(axis=1), columns = ['Missing'])
    # Concatenated missing_values dataframe to input_dataframe by axis 1
    Missing = pd.concat([input_df, missing_values], axis=1)
    return Missing

Missing_values(NHL_Goalies)
```

```python
#Answer 5 here: please comments
# Creating dataframe of goalies without any injuries by searching for null values in Injuries column
NHL_Injured = NHL_Goalies[NHL_Goalies.Injuries.isnull()]

# Renaming incorrectly named columns
GAA_ = GAA_.rename(index=str, columns={"Last Name": "First Name", "First Name": "Last Name"})

# Merging based on 3 values which should be unique: First Name, Last Name and Team
mergedDF_Columns = NHL_Goalies.merge(GAA_, how='inner', on=['First Name', 'Last Name', 'Team'])

# Merging based on index - Requires that the sorted order of the tables are correct prior to merge
mergedDF_Index = pd.merge(NHL_Injured, GAA_, how='inner', left_index=True, right_index=True)

# Keeping First Name, Last Name, Team, Cntry, SV%, GA, GAA, TOI
mergedDF_Columns_subset = mergedDF_Columns[['First Name', 'Last Name', 'Team', 'Cntry', 'SV%_x', 'GA_x', 'GAA_x', 'TOI']]

# Group by country and aggregate means, mins and maxes for all columns
mergedDF_Columns_subset.groupby('Cntry').agg(['mean', 'min', 'max'])

# Group by country and aggregate mean, min and max for TOI
mergedDF_Columns_subset.groupby('Cntry').TOI.agg(['mean', 'min', 'max']).to_csv(path_or_buf ='TOI_2018.csv')
```

# Q6) Write a function(s)  
## that subsets a dataframe by removing rows that are WITHIN the InterQuartile Region of a specified column
<br> </br>
### The function takes as input 
##### A: Pandas_Dataframe (For example use NHL_GoaliesDF)
##### B. Column name
### Add a condition that returns an error if the Column Name chosen is not:
##### A. Numeric Dtype
##### B. Does not exist in the Numeric Columns
### The output is a DataFrame that removes rows outside of the specified columns IQR
<br> </br>
<font color='red'>**You can use smaller helper functions **<font>
<br> </br>
See --> <href>=https://en.wikipedia.org/wiki/Interquartile_range</href>
#hint: There are several pandas subfunctions that can be useful

```python
#Answer Question 6 here
import sys
import pandas.api.types as ptypes

def InterQuartileData(df, ColumnName):
    # Checking if ColumnName exists in the dataframe's column names
    if ColumnName not in df.columns.values:
        # If column name does not exist in dataframe column names, exit function and produce error message
        sys.exit('The column {} does not exist in the user inputted dataframe.'.format(ColumnName))
    if not ptypes.is_numeric_dtype(df[ColumnName]):
        sys.exit('The column {} is not a numeric Dtype.'.format(ColumnName))
    
    # Finding bottom interquartile range
    BottomQuantile = df[ColumnName].quantile(0.25)
    
    # Finding top interquartile range
    TopQuantile = df[ColumnName].quantile(0.75)
    
    # Finding values outside of IQR
    OutsideOfQuantile = df[(df[ColumnName]<BottomQuantile) | (df[ColumnName]>TopQuantile)]
    
    # Finding values inside of IQR
    InsideOfQuantile = df[(df[ColumnName]>=BottomQuantile) & (df[ColumnName]<=TopQuantile)]

    return OutsideOfQuantile, InsideOfQuantile

InterQuartileData(NHL_Goalies, 'GP')
```
